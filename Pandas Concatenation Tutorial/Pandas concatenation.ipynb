```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dir_tutorial = 'C:\\Users\\mpinkert\\Downloads\\Pandas Concatenation Tutorial'
north_america = pd.read_csv(dir_tutorial + '/north_america_2000_2010.csv', index_col=0)
south_america = pd.read_csv(dir_tutorial + '/south_america_2000_2010.csv', index_col=0)
north_america.head()
```

```python
north_america.plot()
```

```python
north_america.transpose().plot(title='Average Labor Hours Per Year')
plt.show()
```

```python
south_america.transpose().plot(title='Average Labor Hours Per Year')
plt.show()
```

```python
americas = pd.concat([north_america, south_america])
americas
```

```python
americas_dfs = [americas]

for year in range(2011, 2016):
    filename = dir_tutorial + "/americas_{}.csv".format(year)
    df = pd.read_csv(filename, index_col=0)
    americas_dfs.append(df)
    
americas_dfs[1]
```

```python
americas = pd.concat(americas_dfs, axis=1)
americas.index.names = ['Country']
americas
```

```python
americas.transpose().plot(title='Average labor hours per year')
plt.show()
```

```python
asia = pd.read_csv(dir_tutorial + '/asia_2000_2015.csv', index_col=0)
asia
```

```python
europe = pd.read_csv(dir_tutorial + '/europe_2000_2015.csv', index_col=0)
europe.head()
```
